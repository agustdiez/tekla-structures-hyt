```python
from markitdown import MarkItDown
from openai import OpenAI

```

```python
ruta = "\\10.10.0.30\Civil\03 - STD TEKLA\docs\manuales\TE_USG_420_en_Template_editor_user_guide.pdf"
```

```python
md = MarkItDown(
    enable_plugins=True,
    llm_client=OpenAI(),
    llm_model="gpt-4o"
)

result = md.convert(ruta)

with open(r"\\10.10.0.30\Civil\03 - STD TEKLA\docs\manuales\salida.md", "w", encoding="utf-8") as f:
    f.write(result.text_content)
```
